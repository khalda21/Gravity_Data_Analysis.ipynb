#  Gravity Data Analysis Project

## Objective
Analyze gravity measurements with depth and visualize the relationship.

## Tools
- Python
- Pandas
- Matplotlib

## Data Description
- Depth (m)
- Gravity (m/sÂ²)

```python
import pandas as pd

data = {
    "Depth": [10, 20, 30, 40, 50],
    "Gravity": [9.78, 9.80, 9.76, 9.79, 9.77]
}

df = pd.DataFrame(data)
df
```

```python
df.head()
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt

plt.plot(df["Depth"], df["Gravity"])
plt.xlabel("Depth")
plt.ylabel("Gravity")
plt.title("Gravity vs Depth")
plt.show()
```

## Conclusion
The analysis shows how gravity values vary with depth.
